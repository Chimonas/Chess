```python
class Piece:
    
    def __init__(self, starting_position, name, white):
        self.position = starting_position
        self.name = name
        self.white = white
 

class King(Piece):
    def __init__(self, white):
        super().__init__(self.calculate_starting_position(white),'K',white)
    def calculate_starting_position(self,white):
        if (white):
            return (0,4)
        else:
            return (7,4)    

class Queen(Piece):
    def __init__(self, white):
        super().__init__(self.calculate_starting_position(white),'Q',white)
    def calculate_starting_position(self,white):
        if (white):
            return (0,3)
        else:
            return (7,3) 
        
class Bishop(Piece):
    def __init__(self, white, index):
        super().__init__(self.calculate_starting_position(white,index),'B',white)
    def calculate_starting_position(self,white,index):
        if (index == 0):
            y = 2
        else:
            y = 5
        if (white):
            return (0,y)
        else:
            return (7,y) 

class Knight(Piece):
    def __init__(self, white, index):
        super().__init__(self.calculate_starting_position(white,index),'A',white)
    def calculate_starting_position(self,white,index):
        if (index == 0):
            y = 1
        else:
            y = 6
        if (white):
            return (0,y)
        else:
            return (7,y) 

class Rook(Piece):
    def __init__(self, white, index):
        super().__init__(self.calculate_starting_position(white,index),'R',white)
    def calculate_starting_position(self,white,index):
        if (index == 0):
            y = 0
        else:
            y = 7
        if (white):
            return (0,y)
        else:
            return (7,y) 
        
class Pawn(Piece):
    def __init__(self, white, index):
        super().__init__(self.calculate_starting_position(white,index),'P',white)
    def calculate_starting_position(self,white,index):
        if (white):
            return (1,index)
        else:
            return (6,index) 
        
class ChessBoard:

    def __init__(self):
        import numpy as np
        self.board = np.chararray([8,8])
        self.alive_pieces = []
        
        self.make_pieces(True)
        self.make_pieces(False)
        self.put_pieces()

    
    def make_pieces(self,white):
        self.alive_pieces.append(King(white))
        self.alive_pieces.append(Queen(white))
        for i in range(2):
            self.alive_pieces.append(Bishop(white,i))
            self.alive_pieces.append(Rook(white,i))
            self.alive_pieces.append(Knight(white,i))
        for i in range(8):
            self.alive_pieces.append(Pawn(white,i))
            
    def put_pieces(self):
        self.board.fill('O')
        for piece in self.alive_pieces:
            self.board[piece.position] = piece.name
        
    def print_board(self):
        for i in range(8):
            line = ""
            for j in range(8):
                line += self.board[i][j].decode("utf-8")  + " "
            print(line)
        

            
board = ChessBoard()
board.print_board()
```
